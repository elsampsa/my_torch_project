```python
green = img[:,:,1]
```

```python
plt.imshow(green, 'gray')
```

```python
img = green
set_to=1
window_size=35
threshold=10
# think this as a mean thresholding 
img_ = cv2.adaptiveThreshold(
    img, 
    set_to, 
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY, 
    window_size, 
    threshold
)
# areas with high values (= nothing) are set to 1
filt = (img_ < 1) # everything that was not marked (as low values ~ 0 mean stuff)
```

```python
plt.imshow(filt,'binary')
```
