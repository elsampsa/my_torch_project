```python
def getPointer(frac):
    """0 == 12:00 0'clock, 1 == 24:00 o'clock
    """
    rad=frac*2*np.pi
    return np.sin(rad), np.cos(rad)
```

```python
print(getPointer(0.25))
```

```python
def genGauge(size, frac, val = 1):
    """Returns gauge bw image & a mask designating the pointer
    """
    mx = int(size*0.9/2)
    cent = (size//2, size//2)
    
    img = np.zeros((size,size), dtype=np.float)
    mask = img.copy()
    cv2.circle(img, cent, mx, val, 2)
    
    tup = getPointer(frac)
    
    dx = int(tup[0]*mx) + cent[0]
    dy = size - (int(tup[1]*mx) + cent[1])
    
    # print(dx, dy)
    
    cv2.arrowedLine(img, cent, (dx, dy), val, 2, 3)    
    # cv2.arrowedLine(mask, cent, (dx, dy), 1, 2, 3)
    cv2.line(mask, cent, (dx, dy), 1, 2, 3)
    
    
    return img, mask
```

```python
# there are quite many ways to randomize this:
# - pointer width (with respect to the circle) & thickness
# - circle radius (spanning the whole image or not)
# - circle & pointer different intensities
# - pointer "decoration" (arrow vs. a plain line)
```

```python
def modImage(img):
    """Add some random blur, intensity, etc. effects
    """
    r = np.random.rand(img.shape[0], img.shape[1]) * 0.5
    img = img + r
    img[img > 1.0] = 1.0
    img = cv2.GaussianBlur(img, (5, 5), 1)
    return img
```

```python
def writeImage(filename, img):
    """Write a 0->1 scaled images as 0->255 uint8 inverted image
    """
    img = (1. - img) * 255
    cv2.imwrite(filename, img.astype(np.uint8))
```

```python
def readImage(filename):
    """Read uint8 grayscale (0->255) image, invert and convert to 0->1
    """
    img = cv2.imread(filename)
    return (255 - img) / 255
```

```python
n = 100 # samples
r = np.random.rand(n, 2)
img_dir = os.path.join("data", "imgs")
mask_dir = os.path.join("data", "masks")
for i, (reading, intensity) in enumerate(r):
    print("creating", i)
    img, mask = genGauge(120, reading, val = intensity)
    img_fname = os.path.join(img_dir, str(i)+".png")
    mask_fname = os.path.join(mask_dir, str(i)+".png")
    writeImage(img_fname, img)
    writeImage(mask_fname, mask)
```
